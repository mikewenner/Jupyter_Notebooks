```python
#imports
import pandas as pd
from pathlib import Path
%matplotlib inline
```

```python
# Read CSV and format on import
csvpath=Path("crypto_data.csv")
crypto_df = pd.read_csv(csvpath, index_col="data_date", parse_dates=True, infer_datetime_format=True)
crypto_df.head()
```

```python
# Drop all columns except cryptocurrency and data_priceUsd
crypto_df.drop(columns=["data_time", "timestamp"], inplace=True)
crypto_df.head()
```

```python
#Group data by cryptocurrency and plot on the same chart
crypto_plot = crypto_df.groupby("cryptocurrency")["data_priceUsd"].plot(legend=True)
```

```python
#find min max
crypto_min = crypto_df.groupby("cryptocurrency")["data_priceUsd"].min()
crypto_min
```

```python
crypto_max = crypto_df.groupby("cryptocurrency")["data_priceUsd"].max()
crypto_max
```

```python
#find average
crypto_avg = round(crypto_df.groupby("cryptocurrency")["data_priceUsd"].mean(), 2)
crypto_avg
```
