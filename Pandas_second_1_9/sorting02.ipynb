```python
import pandas as pd
```

```python
#given dataset
painting_df = pd.DataFrame(
    [
        {"Painting": "Mona Lisa (Knockoff)", "Price": 25, "Popularity": "Very Popular"},
        {"Painting": "Van Gogh (Knockoff)", "Price": 20, "Popularity": "Popular"},
        {"Painting": "Starving Artist", "Price": 10, "Popularity": "Average"},
        {"Painting": "Toddler Drawing", "Price": 1, "Popularity": "Not Popular"},
    ]
)
painting_df
```

```python
# Sort ascending (default) by price
painting_df.sort_values("Price", inplace=True)
painting_df
```

```python
# Sort descending by price
painting_df.sort_values("Price", ascending=False, inplace=True)
painting_df
```

```python
#Sorting the index in dec order
painting_df.sort_index(ascending=False, inplace=True)
painting_df
```

```python
# Set the price as the index
painting_df.set_index("Price", inplace=True)
painting_df
```

```python
# Sort the index in descending order
painting_df.sort_index(ascending=False, inplace=True)
painting_df
```
