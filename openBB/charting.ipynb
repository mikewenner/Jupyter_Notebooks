```python
from openbb_terminal.sdk import openbb

%matplotlib inline
```

```python
# Look at a plot of SPY monthly

spy_monthly = openbb.stocks.load(
        symbol = 'SPY',
        start_date = '1993-11-01',
        monthly = True)
spy_monthly.head()
```

```python
# Now plot using a candle chart
openbb.stocks.candle(
    data = spy_monthly,
    asset_type = 'SPY - Monthly Chart from November, 1993',
    symbol = ''
)
```

```python
#Quick Daily chart for the past 3 years

openbb.stocks.candle("SPY")
```

```python
# look at 15min chart

openbb.stocks.candle("QQQ", interval=15)
```

```python
#Looking at vwap
spy_vwap = openbb.ta.vwap(data = openbb.stocks.load('SPY', interval = 15))
spy_vwap
```

```python
# Putting VWAP on a 15min chart
openbb.ta.vwap_chart(
    data = openbb.stocks.load('SPY', interval = 15))
```

```python
openbb.ta.bbands(
    data = openbb.stocks.load('SPY', interval = 15), window = 15, n_std = 2, mamode = "ema")
```

```python
openbb.ta.bbands_chart(
    data = openbb.stocks.load('SPY', interval = 15), window = 15, n_std = 2, mamode = "ema")
```
