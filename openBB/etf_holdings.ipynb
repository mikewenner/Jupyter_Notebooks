```python
from openbb_terminal.sdk import openbb
```

```python
#look at the holding of an etf --> DIA ***returned sorted by percent of the ETF*****
openbb.etf.holdings("DIA")
```

```python
#use a screener on etf holdings
#bring in the etf to datatframe called symbols
symbols = openbb.etf.holdings('DIA')

#create a list of all the symbols from the DIA ETF, dropping any index that may have N/A
dia_symbols = list(symbols.index.drop(['N/A']))

#bring in valuation data with ca.screener & combine with dia symbols into a dataframe called dia_valuation 
#here if had a list of stocks wanted to compare can change the similar variable, current is just DIA to DIA to see all DIA data
dia_valuation = openbb.stocks.ca.screener(similar = dia_symbols, data_type = 'valuation')  #https://docs.openbb.co/sdk/reference/stocks/ca/screener

dia_valuation = dia_valuation.sort_values(by = ['Price'], ascending = False)#.convert_dtypes() --> changes the market cap column
dia_valuation
#dia_valuation.head(5)
```
