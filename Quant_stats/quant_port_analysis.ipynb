```python
start = '2013-02-13'
end = '2023-02-13'

symbols = ["SPY", "COST", "AVGO", "APD", "AMGN", "CVX", "AAPL", "DIS",
          "BAC", "GS", "SBUX", "COF", "ETN", "XOM", "HD", "MA",
          "QQQ", "IWM", "JNJ", "JPM", "MCD", "MRK", "MSFT", "NVDA", "NKE",
          "PEP", "PFE", "PG", "MO", "BUD", "SHW", "TSCO", "UNH", "VZ", "V", "WMT"]

data = openbb.economy.index(symbols, start_date=start, end_date=end) #.economy.index pulls close price
data
```

```python
data["Date"] = pd.to_datetime(data.index).date
data = data.set_index("Date")
data
```

```python
spy_returns = data["SPY"].pct_change()
spy_returns
```

```python
cost_returns = data["COST"].pct_change()
avgo_returns = data["AVGO"].pct_change()
apd_returns = data["APD"].pct_change()
amgn_returns = data["AMGN"].pct_change()
cvx_returns = data["CVX"].pct_change()
aapl_returns = data["AAPL"].pct_change()
dis_returns = data["DIS"].pct_change()
bac_returns = data["BAC"].pct_change()
gs_returns = data["GS"].pct_change()
sbux_returns = data["SBUX"].pct_change()
cof_returns = data["COF"].pct_change()
etn_returns = data["ETN"].pct_change()
xom_returns = data["XOM"].pct_change()
hd_returns = data["HD"].pct_change()
ma_returns = data["MA"].pct_change()
qqq_returns = data["QQQ"].pct_change()
iwm_returns = data["IWM"].pct_change()
jnj_returns = data["JNJ"].pct_change()
jpm_returns = data["JPM"].pct_change()
mcd_returns = data["MCD"].pct_change()
mrk_returns = data["MRK"].pct_change()
msft_returns = data["MSFT"].pct_change()
nvda_returns = data["NVDA"].pct_change()
nke_returns = data["NKE"].pct_change()
pep_returns = data["PEP"].pct_change()
pfe_returns = data["PFE"].pct_change()
pg_returns = data["PG"].pct_change()
mo_returns = data["MO"].pct_change()
bud_returns = data["BUD"].pct_change()
shw_returns = data["SHW"].pct_change()
tsco_returns = data["TSCO"].pct_change()
unh_returns = data["UNH"].pct_change()
vz_returns = data["VZ"].pct_change()
v_returns = data["V"].pct_change()
wmt_returns = data["WMT"].pct_change()
```

```python
#weight the portfolio (0.0285 = equal wt for 35 stocks)
#weight the portfolio
portfolio = (
    (cost_returns* .0059) + (avgo_returns*.0088) + (apd_returns*.0066) + (amgn_returns*.0029) + (cvx_returns*.0191) + 
    (aapl_returns* .0125) + (dis_returns*.1366) + (bac_returns*.1109) + (gs_returns*.0044) + (sbux_returns*.0389) + 
    (cof_returns* .0132) + (etn_returns*.0264) + (xom_returns*.0345) + (hd_returns*.0066) + (ma_returns*.0125) +
    (qqq_returns* .0800) + (iwm_returns*.0352) + (jnj_returns*.0184) + (jpm_returns*.0228) + (mcd_returns*.0330) + 
    (mrk_returns* .0081) + (msft_returns*.0213) + (nvda_returns*.1043) + (nke_returns*.0264) + (pep_returns*.0110) + 
    (pfe_returns* .0404) + (pg_returns*.0125) + (mo_returns*.0162) + (bud_returns*.0360) + (shw_returns*.0073) + 
    (tsco_returns* .0492) + (unh_returns*.0066) + (vz_returns*.0235) + (v_returns*.0081) + (wmt_returns*.0132)
                )
portfolio.head()
```

```python
#metrics for the weighted portfolio vs spy(benchmark)
qs.reports.metrics(portfolio, spy_returns, mode="full")
```
