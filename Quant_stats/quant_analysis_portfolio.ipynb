```python
#use adjust close to calc daily returns
spy_returns = data_spy["Adj Close"].pct_change()
cost_returns = data_cost["Adj Close"].pct_change()
avgo_returns = data_avgo["Adj Close"].pct_change()
apd_returns = data_apd["Adj Close"].pct_change()
amgn_returns = data_amgn["Adj Close"].pct_change()
cvx_returns = data_cvx["Adj Close"].pct_change()
aapl_returns = data_aapl["Adj Close"].pct_change()
dis_returns = data_dis["Adj Close"].pct_change()
bac_returns = data_bac["Adj Close"].pct_change()
gs_returns = data_gs["Adj Close"].pct_change()
sbux_returns = data_sbux["Adj Close"].pct_change()
cof_returns = data_cof["Adj Close"].pct_change()
etn_returns = data_etn["Adj Close"].pct_change()
xom_returns = data_xom["Adj Close"].pct_change()
hd_returns = data_hd["Adj Close"].pct_change()
ma_returns = data_ma["Adj Close"].pct_change()
qqq_returns = data_qqq["Adj Close"].pct_change()
iwm_returns = data_iwm["Adj Close"].pct_change()
jnj_returns = data_jnj["Adj Close"].pct_change()
jpm_returns = data_jpm["Adj Close"].pct_change()
mcd_returns = data_mcd["Adj Close"].pct_change()
mrk_returns = data_mrk["Adj Close"].pct_change()
msft_returns = data_msft["Adj Close"].pct_change()
nvda_returns = data_nvda["Adj Close"].pct_change()
nke_returns = data_nke["Adj Close"].pct_change()
pep_returns = data_pep["Adj Close"].pct_change()
pfe_returns = data_pfe["Adj Close"].pct_change()
pg_returns = data_pg["Adj Close"].pct_change()
mo_returns = data_mo["Adj Close"].pct_change()
bud_returns = data_bud["Adj Close"].pct_change()
shw_returns = data_shw["Adj Close"].pct_change()
tsco_returns = data_tsco["Adj Close"].pct_change()
unh_returns = data_unh["Adj Close"].pct_change()
vz_returns = data_vz["Adj Close"].pct_change()
v_returns = data_v["Adj Close"].pct_change()
wmt_returns = data_wmt["Adj Close"].pct_change()
```

```python
#weight the portfolio
portfolio = (
    (cost_returns* .0059) + (avgo_returns*.0088) + (apd_returns*.0066) + (amgn_returns*.0029) + (cvx_returns*.0191) + 
    (aapl_returns* .0125) + (dis_returns*.1366) + (bac_returns*.1109) + (gs_returns*.0044) + (sbux_returns*.0389) + 
    (cof_returns* .0132) + (etn_returns*.0264) + (xom_returns*.0345) + (hd_returns*.0066) + (ma_returns*.0125) +
    (qqq_returns* .0800) + (iwm_returns*.0352) + (jnj_returns*.0184) + (jpm_returns*.0228) + (mcd_returns*.0330) + 
    (mrk_returns* .0081) + (msft_returns*.0213) + (nvda_returns*.1043) + (nke_returns*.0264) + (pep_returns*.0110) + 
    (pfe_returns* .0404) + (pg_returns*.0125) + (mo_returns*.0162) + (bud_returns*.0360) + (shw_returns*.0073) + 
    (tsco_returns* .0492) + (unh_returns*.0066) + (vz_returns*.0235) + (v_returns*.0081) + (wmt_returns*.0132)
                )
portfolio.head()
```

```python
#metrics for the weighted portfolio vs spy(benchmark)
qs.reports.metrics(portfolio, spy_returns, mode="full")
```
