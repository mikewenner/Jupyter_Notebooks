```python
# Import libraries and dependencies
import pandas as pd
from pathlib import Path
%matplotlib inline
```

```python
#Read CSV in as DataFrame
csvpath = Path("sp500_companies.csv")
sp500_df = pd.read_csv(csvpath)
sp500_df.head()
```

```python
#Calculate Sector Counts
# Count the frequency of each sector from the list of companies
sector_count = sp500_df["Sector"].value_counts()
sector_count
```

```python
#Plot Sector Counts
# Plot a pie chart from the distribution of company sectors
sector_count.plot(kind="pie")
```

```python
#Filter By Top 20 Market Cap Companies
#Create a subset DataFrame by selecting the Symbol and Market Cap columns. 
#Use the nlargest() function on the subset DataFrame to return the top 20 rows of the Market Cap column.

market_cap = sp500_df.loc[:, ["Symbol", "Market Cap"]]
market_cap.head()
```

```python
# Set the 'Symbol' as the index
market_cap.set_index(market_cap["Symbol"], inplace=True)
market_cap.head()
```

```python
# Drop the extra 'Symbol' column
#market_cap.drop(columns=["Symbol"], inplace=True)
#market_cap.head()
market_cap
```

```python
# Filter down to 20 companies with the largest market caps
market_cap_top20 = market_cap.nlargest(20, "Market Cap")
market_cap_top20
```
