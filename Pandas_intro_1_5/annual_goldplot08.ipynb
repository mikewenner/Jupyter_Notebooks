```python
import pandas as pd
from pathlib import Path
%matplotlib inline
```

```python
#1. Read in CSV
csvpath = Path("annual_gold.csv")
annual_gold = pd.read_csv(csvpath)
annual_gold.head()
```

```python
# Plot a line chart of the data without setting the 'Date' column as the index
annual_gold.plot()
```

```python
# Convert date strings into datetime objects and set the datetime as the index
annual_gold.set_index(pd.to_datetime(annual_gold["Date"], infer_datetime_format=True), inplace=True)
annual_gold.head()
```

```python
# Drop the extra date column
annual_gold.drop(columns=["Date"], inplace=True)
annual_gold.head()
```

```python
#. Plot Data with Datetime Index
annual_gold.plot()
```

```python
## Plot a bar chart of the data
annual_gold.plot(kind = "bar")
```

```python
#Plot Bar Chart with Increased Figure Size
annual_gold.plot(kind = "bar", figsize=(15, 5))
```
