```python
# initial imports
import pandas as pd
from pathlib import Path
```

```python
# set the file path
# create a Pandas dataframe from a csv file

csvpath = Path('amd_stock_data.csv')
amd_stock_data_dataframe = pd.read_csv(csvpath)
```

```python
# get the first 10 rows from the dataframe
amd_stock_data_dataframe.head(10)
```

```python
# set column names
# recreate the dataframe
# add columns names "Date", "Close", "Volume", "Open", "High", "Low"

amd_stock_data_dataframe = pd.read_csv(csvpath, header = None)
column_names = ["Date", "Close", "Volume", "Open", "High", "Low"]
amd_stock_data_dataframe.columns = column_names
```
