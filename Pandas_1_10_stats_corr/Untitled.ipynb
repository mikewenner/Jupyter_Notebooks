```python
# Import libraries and dependencies
import pandas as pd
from pathlib import Path
%matplotlib inline
```

```python
# Set file path
nflx_data = Path("nflx.csv")
```

```python
# Read the CSV into a DataFrame and set the 'date' column as a datetime index
nflx_df = pd.read_csv(nflx_data, index_col="date", infer_datetime_format=True, parse_dates=True)
```

```python
#sort the index
nflx_df.sort_index(inplace=True)
nflx_df.head()
```

```python
# Plot DataFrame
nflx_df.plot()
```

```python
# Calculate 20-day Simple Moving Average for NFLX Closing Prices
nflx_sma20 = nflx_df.rolling(window=20).mean()
nflx_sma20
```

```python
# Calculate 50-day Simple Moving Average for NFLX Closing Prices
nflx_sma50 = nflx_df.rolling(window=50).mean()
nflx_sma50
```

```python
# Calculate 100-day Simple Moving Average for NFLX Closing Prices
nflx_sma100 = nflx_df.rolling(window=100).mean()
nflx_sma100
```

```python
#Overlay Daily Closing Prices of NFLX and SMAs
ax = nflx_df.plot(figsize=(25, 10))
nflx_sma20.plot(ax=ax)
nflx_sma50.plot(ax=ax)
nflx_sma100.plot(ax=ax)

# Set the legend of the figure
ax.legend(["Original", "SMA20", "SMA50", "SMA100"]);
```
