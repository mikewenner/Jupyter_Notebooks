```python
import pandas as pd
from pathlib import Path
%matplotlib inline
```

```python
#set path read in nflx csv
nflx_path = Path("nflx.csv")
nflx_df = pd.read_csv(nflx_path, index_col="date", parse_dates=True, infer_datetime_format=True)
#set index column to date

nflx_df.sort_index(inplace=True)
nflx_df.head()
```

```python
nflx_df.plot()
```

```python
# Calculate 20-day Simple Moving Average for NFLX Closing Prices
nflx_20sma = nflx_df.rolling(window=20).mean()
nflx_20sma
```

```python
# Calculate 50 & 100-day Simple Moving Average for NFLX Closing Prices
nflx_50sma = nflx_df.rolling(window=50).mean()
nflx_100sma = nflx_df.rolling(window=100).mean()
```

```python
#Overlay Daily Closing Prices of NFLX and SMAs
ax = nflx_df.plot(figsize=(20, 10), title="NFLX Daily")

nflx_20sma.plot(ax=ax, linestyle="dotted")
nflx_50sma.plot(ax=ax)
nflx_100sma.plot(ax=ax)

ax.legend(["close", "sma20", "sma50", "sma100"]);
```
