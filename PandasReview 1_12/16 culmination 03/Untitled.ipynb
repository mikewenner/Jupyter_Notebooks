```python
# Import dependencies
import pandas as pd
from pathlib import Path
```

```python
csvpath = Path("world_happiness.csv")
world_happiness_df = pd.read_csv(csvpath, low_memory=False)
world_happiness_df.head()
```

```python
# Set new index to Country
world_happiness_df.set_index("Country", inplace=True)
world_happiness_df.head()
```

```python
# Sort the new dataframe alphabetically
world_happiness_df.sort_index(inplace=True)
world_happiness_df.head()
```

```python
correlation = world_happiness_df.corr()
correlation
```

```python
import seaborn as sns
sns.heatmap(correlation, vmin=-1, vmax=1)

#whisker hih & low best corr to happiness
```

```python
world_happy_avg = world_happiness_df.mean()
world_happy_avg
```
