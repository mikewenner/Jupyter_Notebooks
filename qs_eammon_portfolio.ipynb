```python
#use adjust close to calc daily returns
spy_returns = data_spy["Adj Close"].pct_change()
abbv_returns = data_abbv["Adj Close"].pct_change()
acn_returns = data_acn["Adj Close"].pct_change()
apd_returns = data_apd["Adj Close"].pct_change()
amgn_returns = data_amgn["Adj Close"].pct_change()
apo_returns = data_apo["Adj Close"].pct_change()
aapl_returns = data_aapl["Adj Close"].pct_change()
t_returns = data_t["Adj Close"].pct_change()
bac_returns = data_bac["Adj Close"].pct_change()
avgo_returns = data_avgo["Adj Close"].pct_change()
ko_returns = data_ko["Adj Close"].pct_change()
cci_returns = data_cci["Adj Close"].pct_change()
etn_returns = data_etn["Adj Close"].pct_change()
xom_returns = data_xom["Adj Close"].pct_change()
hd_returns = data_hd["Adj Close"].pct_change()
ibm_returns = data_ibm["Adj Close"].pct_change()
qqq_returns = data_qqq["Adj Close"].pct_change()
rpv_returns = data_rpv["Adj Close"].pct_change()
jnj_returns = data_jnj["Adj Close"].pct_change()
jpm_returns = data_jpm["Adj Close"].pct_change()
mdt_returns = data_mdt["Adj Close"].pct_change()
mrk_returns = data_mrk["Adj Close"].pct_change()
msft_returns = data_msft["Adj Close"].pct_change()
nee_returns = data_nee["Adj Close"].pct_change()
nke_returns = data_nke["Adj Close"].pct_change()
pep_returns = data_pep["Adj Close"].pct_change()
pfe_returns = data_pfe["Adj Close"].pct_change()
pg_returns = data_pg["Adj Close"].pct_change()
pld_returns = data_pld["Adj Close"].pct_change()
o_returns = data_o["Adj Close"].pct_change()
shw_returns = data_shw["Adj Close"].pct_change()
kre_returns = data_kre["Adj Close"].pct_change()
unh_returns = data_unh["Adj Close"].pct_change()
vz_returns = data_vz["Adj Close"].pct_change()
v_returns = data_v["Adj Close"].pct_change()
wmt_returns = data_wmt["Adj Close"].pct_change()
```

```python
#weight the portfolio
portfolio = (
    (abbv_returns* .0059) + (acn_returns*.0088) + (apd_returns*.0066) + (amgn_returns*.0029) + (apo_returns*.0191) + 
    (aapl_returns* .0125) + (t_returns*.1366) + (bac_returns*.1109) + (avgo_returns*.0044) + (ko_returns*.0389) + 
    (cci_returns* .0132) + (etn_returns*.0264) + (xom_returns*.0345) + (hd_returns*.0066) + (ibm_returns*.0125) +
    (qqq_returns* .0800) + (rpv_returns*.0352) + (jnj_returns*.0184) + (jpm_returns*.0228) + (mdt_returns*.0330) + 
    (mrk_returns* .0081) + (msft_returns*.0213) + (nee_returns*.0235) + (nke_returns*.0264) + (pep_returns*.0110) + 
    (pfe_returns* .0404) + (pg_returns*.0125) + (pld_returns*.0162) + (o_returns*.0360) + (shw_returns*.0073) + 
    (kre_returns* .0492) + (unh_returns*.0066) + (vz_returns*.1043) + (v_returns*.0081) + (wmt_returns*.0132)

            )
portfolio.head()
```

```python
#metrics for the weighted portfolio vs spy(benchmark)
qs.reports.metrics(portfolio, spy_returns, mode="full")
```
