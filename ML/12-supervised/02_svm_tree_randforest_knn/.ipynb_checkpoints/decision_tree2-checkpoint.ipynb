```python
# Initial imports
import pandas as pd
from pathlib import Path
from sklearn import tree
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

# Needed for decision tree visualization
import pydotplus
from IPython.display import Image
```

```python
# Loading data
df_loans = pd.read_csv(Path("sba_loans_encoded.csv"))
df_loans
```

```python
# Define target vector
y = df_loans["Default"].values.reshape(-1,1)
y[:5]
```

```python
# Define features set
X = df_loans.drop(columns=["Default"])
X.head()
```

```python
# Splitting into Train and Test sets
X_train, X_test, y_train, y_test = train_test_split(X,y)
```

# scale the data

```python
# Create the StandardScaler instance
scaler = StandardScaler()
```

```python
# Fit the Standard Scaler with the training data
X_scaler = scaler.fit(X_train)
X_scaler
```

```python
# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

# model fit predict

```python
# Create the decision tree classifier instance
model = tree.DecisionTreeClassifier()
```

```python
# Fit the model
model = model.fit(X_train_scaled, y_train)
```

```python
# Making predictions using the testing data
predictions = model.predict(X_test_scaled)
```

```python
#looking at score on train data --- 1.0 beacuse tree model tend to overfirt on training data as compared to the test score below
model.score(X_train_scaled, y_train)
```

```python
#looking at score on test data
model.score(X_test_scaled, y_test)
```

```python
# Calculating the confusion matrix
cm = confusion_matrix(y_test, predictions)
cm_df = pd.DataFrame(
    cm, index=["Actual 0", "Actual 1"], columns=["Predicted 0", "Predicted 1"]
)

# Calculating the accuracy score
acc_score = accuracy_score(y_test, predictions)
acc_score
```

```python
# Displaying results
print("Confusion Matrix")
display(cm_df)
print(f"Accuracy Score : {acc_score}")
print("Classification Report")
print(classification_report(y_test, predictions))
```

```python
# Create DOT data
dot_data = tree.export_graphviz(
    model, out_file=None, feature_names=X.columns, class_names=["0", "1"], filled=True
)

# Draw graph
graph = pydotplus.graph_from_dot_data(dot_data)

# Show graph
Image(graph.create_png())
```
